```python
# _*_ coding: UTF-8 _*_
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
```

```python
html = urlopen('https://overwatch.op.gg/leaderboards')
html
```

```python
bsObject = BeautifulSoup(html,'html.parser')
bsObject
```

```python
for n in range(0,25):
    td = bsObject.find_all('td',{'class':'ContentCell ContentCell-Player'})
    ad = bsObject.find_all('td',{'class':'ContentCell ContentCell-SkillRating'})
    cd = bsObject.find_all('td',{'class':'ContentCell ContentCell-KD'})
    ID = td[n].select('b')[0].get_text()
    Level = td[n].select('em')[0].get_text()
    Rating = ad[n].get_text().strip()
    KDA = cd[n].select('b')[0].get_text()
    print(n+1,ID,Level,Rating, KDA)
```

```python
with open('crawling.csv', 'w',encoding='UTF-8') as f:
    writer = csv.writer(f)
    writer.writerow(['rank', 'ID', 'Level', 'Rate Score', 'KDA'])
    
    for n in range(0,25):
        td = bsObject.find_all('td',{'class':'ContentCell ContentCell-Player'})
        ad = bsObject.find_all('td',{'class':'ContentCell ContentCell-SkillRating'})
        cd = bsObject.find_all('td',{'class':'ContentCell ContentCell-KD'})
        ID = td[n].select('b')[0].get_text()
        Level = td[n].select('em')[0].get_text()
        Rating = ad[n].get_text().strip()
        KDA = cd[n].select('b')[0].get_text()
        writer.writerow([n+1,ID,Level,Rating, KDA])
```
